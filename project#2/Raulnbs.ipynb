```python
# Now add the Gaussian distribution
pi= np.pi
def sqrt(x): return np.sqrt(x)
def exp(x): return np.exp(x)
```

```python
# Gaussian function
```

```python
def f_gaus(x,mu,sig):
    amp = 1.0/sqrt(2.0 * pi * pow(sig,2))
    arg = pow(x-mu,2)/( 2.0 * pow(sig,2) )
    return amp * exp(-arg)

m0, dm0 = 1, .01
```

```python
# Make a distribution of m values
# Use: https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
Npoints = 1000
m0s = np.random.normal(m0, dm0, Npoints)
```

```python
# Show the data
m0s
```

```python
# Save these data in an external file
filename='data.txt'
np.savetxt(filename,m0s)
```

```python
# Create a histogram from the numpy module
Nbins = 25
hist, bin_edges = np.histogram(m0s,bins=Nbins)
```

```python
plt.hist(m0s, bins=Nbins)  

x=np.arange(m0-4*dm0,m0+4*dm0,dm0/100.)

tmp = f_gaus(x,m0,dm0)
tmp = tmp * max(hist) / max(tmp) 

plt.plot(x,tmp ,color='r')

plt.ylabel('Number of points',size=20)
plt.xlabel('m_0',size=20)

```
